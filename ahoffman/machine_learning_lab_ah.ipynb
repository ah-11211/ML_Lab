```python
df2 = df2.rename(columns = {'Order ID': 'Order.ID'})
df2
```

```python
df = pd.merge(df, df2, how='left', on = ['Order.ID'])
df
```

```python
df['Returned'] = df['Returned'].fillna('No')
```

```python
df = df.drop('Region_y', axis=1)
```

```python
df
```

```python
df = df.rename(columns = {'Region_x': 'Region'})
df['Order.Year'] = pd.DatetimeIndex(df['Order.Date']).year
```

```python
df
```

```python
returned_items = df[df['Returned'] == 'Yes'].reset_index()
returned_items.groupby(['Order.Year']).agg(['sum'])['Profit']
```

```python
customer_returns = returned_items.groupby('Customer.ID').count().reset_index()
customer_returns
```

```python
print(sum(customer_returns['Returned']>1))
print(sum(customer_returns['Returned']>5))
```
